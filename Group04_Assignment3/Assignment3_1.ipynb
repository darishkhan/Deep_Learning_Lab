**Converting Data in trainable format**

```python
import os
import pandas as pd
from sklearn.utils import shuffle #---> shuffles the dataframe
from glob import glob
from PIL import Image
import numpy as np

data_dir = '/content/extracted_data/Data'

def get_image_data_and_labels(base_dir, img_size=(224, 224)):
    image_data = []
    labels = []

    for class_name in os.listdir(base_dir):
        class_dir = os.path.join(base_dir, class_name)

        if os.path.isdir(class_dir):
            images = glob(os.path.join(class_dir, '*.jpg'))

            for img_path in images:
                img = Image.open(img_path).resize(img_size)
                img = np.array(img)

                image_data.append(img)
                labels.append(class_name)

    return image_data, labels

train_images, train_labels = get_image_data_and_labels(os.path.join(data_dir, 'train'))
test_images, test_labels = get_image_data_and_labels(os.path.join(data_dir, 'test'))
val_images, val_labels = get_image_data_and_labels(os.path.join(data_dir, 'val'))

train_df = pd.DataFrame({'image': train_images, 'label': train_labels})
test_df = pd.DataFrame({'image': test_images, 'label': test_labels})
val_df = pd.DataFrame({'image': val_images, 'label': val_labels})

train_df = shuffle(train_df)
test_df = shuffle(test_df)
val_df = shuffle(val_df)

print("Train DataFrame:\n", train_df.head())
print("Test DataFrame:\n", test_df.head())
print("Validation DataFrame:\n", val_df.head())
```

```python
train_df.head()
```

```python


train_df_shuffled = train_df.sample(frac=1).reset_index(drop=True)
test_df_shuffled = test_df.sample(frac=1).reset_index(drop=True)
val_df_shuffled = val_df.sample(frac=1).reset_index(drop=True)

```

**Some samples are grayscale , imputing them to RGB**

```python

import numpy as np
def to_rgb(image):
  if len(image.shape) == 2:  # Grayscale image
    return np.repeat(image[:, :, np.newaxis], 3, axis=2)
  elif image.shape[2] == 1:  # Single-channel image
    return np.repeat(image, 3, axis=2)
  else:
    return image

train_df_shuffled['image'] = train_df_shuffled['image'].apply(to_rgb)
test_df_shuffled['image'] = test_df_shuffled['image'].apply(to_rgb)
val_df_shuffled['image'] = val_df_shuffled['image'].apply(to_rgb)
```

```python
val_df_shuffled.shape
```

**Preprocessing the data**

```python

import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dropout
from tensorflow.keras import layers, regularizers, models




# preprocessing the data
train_images = np.array([np.array(img) for img in train_df_shuffled['image'].tolist()]) # Ensure all images are converted to NumPy arrays with consistent shapes
train_labels = np.array(train_df_shuffled['label'].tolist())
test_images = np.array([np.array(img) for img in test_df_shuffled['image'].tolist()]) # Ensure all images are converted to NumPy arrays with consistent shapes
test_labels = np.array(test_df_shuffled['label'].tolist())
val_images = np.array([np.array(img) for img in val_df_shuffled['image'].tolist()]) # Ensure all images are converted to NumPy arrays with consistent shapes
val_labels = np.array(val_df_shuffled['label'].tolist())

# normalising the values between 0 to 1
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0
val_images = val_images.astype('float32') / 255.0

# converting to numerical values
unique_labels = np.unique(train_labels)
label_mapping = {label: index for index, label in enumerate(unique_labels)}
train_labels_numeric = np.array([label_mapping[label] for label in train_labels])
test_labels_numeric = np.array([label_mapping[label] for label in test_labels])
val_labels_numeric = np.array([label_mapping[label] for label in val_labels])



train_images.shape, test_images.shape, val_images.shape, train_labels_numeric.shape, test_labels_numeric.shape, val_labels_numeric.shape
```

**FCNN without dropout and regularization**

```python
from tensorflow.keras import layers, models, regularizers

model = models.Sequential([
    layers.Input(shape=(224, 224, 3)),  # Input shape of the image (224x224 with 3 color channels)

    # Flattening the 2D image to 1D vector
    layers.Flatten(),

    # Fully connected layers
    layers.Dense(128, activation="relu"),

    layers.Dense(64, activation="relu"),

    # Output layer with softmax activation for multi-class classification
    layers.Dense(5, activation="softmax")
])

# Compiling the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
test_images.shape
```

**Training on the above model**

```python
from tensorflow.keras.callbacks import EarlyStopping

#When validating accuracy falls low, it stops training
early_stopping = EarlyStopping(monitor='val_loss',
                               patience=5,
                               restore_best_weights=True)

model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

history = model.fit(train_images, train_labels_numeric, batch_size=32,
                    epochs=150,
                    validation_data=(val_images, val_labels_numeric),
                    callbacks=[early_stopping])

test_loss, test_acc = model.evaluate(test_images, test_labels_numeric, verbose=2)
print("\nTest accuracy:", test_acc)
```

**Confusion Matrix for the above training**

```python

import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Predict labels for test data
y_pred = model.predict(test_images)
y_pred_classes = np.argmax(y_pred, axis=1)

# Calculate confusion matrix
cm = confusion_matrix(test_labels_numeric, y_pred_classes)

# Plot confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=unique_labels, yticklabels=unique_labels)
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix")
plt.show()
```
